```python
import pandas as pd
import numpy as np

rd = pd.read_csv('rawdata.csv', encoding='cp949')

rd.head()
```

```python
rd_count = rd.groupby(['자치구명']).size()
print(rd_count)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'

rd_count_sorted = rd_count.sort_values(ascending=False)

plt.figure(figsize=(10,6))
ax = rd_count_sorted.plot(kind='bar', color='pink')
plt.title('서울시 거래량(건)')
plt.xlabel('자치구')
plt.ylabel('거래수(건)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()

bars=ax.patches
bars[0].set_color('purple')

plt.show()
```
